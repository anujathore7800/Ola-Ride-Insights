### Missing values Columns Updated

* **Canceled_Rides_by_Customer** → Filled with "Not Applicable" because cancellation reasons are only available for rides cancelled by customers.
* **Canceled_Rides_by_Driver** → Filled with "Not Applicable" because cancellation reasons are only available for rides cancelled by drivers.
* **Incomplete_Rides** → Filled with "Not Applicable" because this field is only relevant for incomplete rides.
* **Incomplete_Rides_Reason** → Filled with "Not Applicable" because reasons exist only when a ride is marked as incomplete.
* **Payment_Method** → Filled with "Not Applicable" because payment details are only available for successfully completed rides.

### Some missing value columns are null because--
* **V_TAT** → Kept as NULL because turnaround time was not available for rides that were not successfully completed.
* **C_TAT** → Kept as NULL because customer turnaround time could not be measured for certain non-completed rides.
* **Driver_Ratings** → Kept as NULL because ratings are only provided after a successful ride. Filling with 0 would incorrectly suggest a poor rating.
* **Customer_Rating** → Kept as NULL because ratings are not available for cancelled or unsuccessful rides. A value of 0 would distort customer satisfaction analysis.

### Feature Engineering

```python
df['Time']
```

```python
df['Hour'] = df['Time'].apply(lambda x: x.hour)
```

```python
df[['Time', 'Hour']].head()
```

```python
def time_bucket(hour):
    if 0 <= hour < 6:
        return 'Night'
    elif 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    else:
        return 'Evening'

df['Time_Bucket'] = df['Time'].apply(lambda x: time_bucket(x.hour))
```
